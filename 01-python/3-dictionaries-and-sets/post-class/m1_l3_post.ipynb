```python
# Q1:

prices = {"milk": 60, "bread": 40, "eggs": 90, "butter": 120}
# keys, values, items, get
print(prices['milk'])
print(prices.get('milk'))
# print(prices['milks']) // Error
print(prices.get('milks', 0))

items = list(prices)
print(items)
print(items[0][0], items[0][1])

# type
print(type(prices))
print(type(items))
```
