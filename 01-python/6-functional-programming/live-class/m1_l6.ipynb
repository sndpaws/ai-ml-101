```python
masai = [ 100000 ,200000, 40000 ]
mars = [ 120000, 250000, 410000 ]
earth = [ 120000, 250000, 410000, 2000000]

# average salary for masai
# average salary for Mars
# average salary for all the companies
def add(list_of_values):
  answer = 0;
  for value in list_of_values:
    answer += value
  return answer;

def average(total, count):
  return total / count;

# average_salary_for_company ->
def average_salary(salaries):
  total = add(salaries)
  count = len(salaries)
  avg = average(total, count)
  return int(avg);

company1 = average_salary(masai)
company2 = average_salary(mars)
company3 = average_salary(earth)

company_avgs = [company1, company2, company3];
print(average_salary(company_avgs))
# average_salaries_world_wide
# average_salaries_world_wide([company1, company2, company3])
```
