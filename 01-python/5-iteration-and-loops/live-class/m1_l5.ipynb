```python
# for in
# range
# list, set, tuple, dict for loop
# salaries = {'Rahul': 49000, 'Priya': 45000, 'Amit': 75000}
salaries = [
    {'name': 'Rahul', 'salary' : '49000'},
    {'name': 'Priya', 'salary' : '45000'},
    {'name': 'Amit', 'salary' : '75000'},
]

monthly_salary_tax_bar = 600000 / 12;
# list
for emp in salaries:
  if int(emp['salary']) >= monthly_salary_tax_bar:
    print(f'Hi {emp['name']}, please pay tax')
  else:
    print(f'Hi {emp['name']}, you do not have to pay tax')


```

```python
power = 30;

# while, break

# if battery is below 20% then show warning
# if battery is below 2% then close background apps

# code
# Iterating in a Loop
# For [start.. end]
# while condition
while power > 0:
  if power == 20:
    print('Battery is 20%')
    break;
  elif power <= 2:
    print('Closing all apps')
  else:
    print('Phone has power: ', power);
  power -= 1

```
